```python
from matplotlib import pyplot
import numpy as np
import plotly.graph_objects as go

# y' = func(x, y), y(x0) = y0
x0 = 0
y0 = 1
def func(x, y):
    return np.cos(x)

# solution y(x) for y' = func(x, y)
def sol(x):
    return x ** 3 + x ** 2

# plot obj
fig = go.Figure()

# our method
def runge_kutta(x0, y0, h, N):
    xp = [x0]
    yp = [y0]
    for i in range(1, N):
        xn = xp[i - 1]
        yn = yp[i - 1]
        k1 = func(xn, yn)
        k2 = func(xn + h/2, yn + k1 * h / 2)
        k3 = func(xn + h/2, yn + k2 * h / 2)
        k4 = func(xn + h, yn + k2 * h)
        x_new = xn + h
        y_new = yn + h / 6 * (k1 + 2*k2 + 2*k3 + k4)
        xp.append(x_new)
        yp.append(y_new)
    fig.add_trace(go.Scatter(x=xp, y=yp, mode='markers', name="Runge-Kutta")) # adding method plot 
    return xp, yp

h = 0.1
N = 500

xp, yp = runge_kutta(x0, y0, h, N)

t = np.linspace(x0, x0 + h * N, 1000)
fig.add_trace(go.Scatter(x=t, y=sol(t), name="Analytic solution")) # adding real solution plot
fig.show()

eps = 0
for i in range(0, N):
    eps = max(eps, abs(yp[i] - sol(xp[i])))
print("max error = ", eps)
```
